## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Total Schools
total_schools = school_df["School ID"].count()
#Total Students
total_students = student_df["Student ID"].count()
#Total Budget
total_budget = school_df["budget"].sum()
#Average Math Score
average_math_score = student_df['math_score'].mean()
#Average Reading Score
average_read_score = student_df['reading_score'].mean()
#Precentage of Students Passing Math
pre_pass_math = ((student_df['math_score']>=70).sum()/total_students)*100
#Precentage of Students Passing Math
pre_pass_reading = ((student_df['reading_score']>=70).sum()/total_students)*100
#Precentage of Students Passing Overall
pre_pass_overall = ((pre_pass_math+pre_pass_reading)/2)
```

```python
district_overview_df = pd.DataFrame({
    "Total Schools":[total_schools],
    "Total Students":[total_students],
    "Total Budget":[total_budget],
    "Average Math Score":[average_math_score],
    "Average Reading Score":[average_read_score],
    "% Passing Math":[pre_pass_math],
    "% Passing Reading":[pre_pass_reading],
    "% Passing Overall":[pre_pass_overall]
})
district_overview_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score

  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Change index to school name and determine the school type
school_type = school_df.set_index(["school_name"])["type"]
#Total Students
students_by_school = school_data_complete_renamed_df["School Name"].value_counts()
#Total School Budget
school_budget_by_school = school_data_complete_renamed_df.groupby(["School Name"]).mean()["Budget"]
#Per Student Budget
student_budget_by_school = school_budget_by_school/ students_by_school
#Average Math Score
average_math_score_by = school_data_complete_renamed_df.groupby(["School Name"]).mean()["Math Score"]
#Average Reading Score
average_reading_score_by = school_data_complete_renamed_df.groupby(["School Name"]).mean()["Reading Score"]
#% Passing Math
pass_math_by_school =  ((school_data_complete_renamed_df[school_data_complete_renamed_df["Math Score"] > 70].groupby("School Name").count()["Student Name"])/students_by_school) * 100
#% Passing Readings
pass_reading_by_school =  ((school_data_complete_renamed_df[school_data_complete_renamed_df["Reading Score"] > 70].groupby("School Name").count()["Student Name"])/students_by_school) * 100
#Overall Passing Rate (Average of the above two)
overall_passing_rate = (pass_math_by_school + pass_reading_by_school) / 2
```

```python
School_Summary = pd.DataFrame({"School Type":school_type, "Total Students":students_by_school,
                               "Total School Budget":school_budget_by_school, "Per Student Budget":student_budget_by_school,
                                "Average Math Score":average_math_score_by,"Average Reading Score":average_reading_score_by,
                               "% Passing Math":pass_math_by_school,"% Passing Reading":pass_reading_by_school,
                               "Overall Passing Rate":overall_passing_rate})

# Display the data frame
School_Summary
```
