```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_df = pd.read_csv(school_data_to_load)
student_df = pd.read_csv(student_data_to_load)

student_df.head()
school_df.head()
```

```python
# Combine the data into a single dataset
school_data_complete_df = pd.merge(student_df, school_df, on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
school_data_complete_renamed_df = school_data_complete_df.rename(columns = {"student_name": "Student Name", "gender": "Gender", "grade": "Grade",
                                           "school_name": "School Name", "reading_score": "Reading Score",
                                           "math_score": "Math Score", "type": "Type", "size": "Size", "budget":
                                           "Budget"})
school_data_complete_renamed_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Total Schools
total_schools = school_df["School ID"].count()
#Total Students
total_students = student_df["Student ID"].count()
#Total Budget
total_budget = school_df["budget"].sum()
#Average Math Score
average_math_score = student_df['math_score'].mean()
#Average Reading Score
average_read_score = student_df['reading_score'].mean()
#Precentage of Students Passing Math
pre_pass_math = ((student_df['math_score']>=70).sum()/total_students)*100
#Precentage of Students Passing Math
pre_pass_reading = ((student_df['reading_score']>=70).sum()/total_students)*100
#Precentage of Students Passing Overall
pre_pass_overall = ((pre_pass_math+pre_pass_reading)/2)
```

```python
district_overview_df = pd.DataFrame({
    "Total Schools":[total_schools],
    "Total Students":[total_students],
    "Total Budget":[total_budget],
    "Average Math Score":[average_math_score],
    "Average Reading Score":[average_read_score],
    "% Passing Math":[pre_pass_math],
    "% Passing Reading":[pre_pass_reading],
    "% Passing Overall":[pre_pass_overall]
})
district_overview_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Change index to school name and determine the school type
school_type = school_df.set_index(["school_name"])["type"]
#Total Students
by_school = school_data_complete_renamed_df.groupby("School Name")
students_by_school = by_school['Student ID'].count()
#Total School Budget
school_budget_by_school = school_data_complete_renamed_df.groupby(["School Name"]).mean()["Budget"]
#Per Student Budget
student_budget_by_school = school_budget_by_school/ students_by_school
#Average Math Score
average_math_score_by = school_data_complete_renamed_df.groupby(["School Name"]).mean()["Math Score"]
#Average Reading Score
average_reading_score_by = school_data_complete_renamed_df.groupby(["School Name"]).mean()["Reading Score"]
#% Passing Math
school_math = school_data_complete_renamed_df.loc[school_data_complete_renamed_df['Math Score'] >= 70, :].groupby("School Name")
math_counts = school_math['Student ID'].count()
math_rates = (math_counts/students_by_school)*100
#% Passing Readings
school_read = school_data_complete_renamed_df.loc[school_data_complete_renamed_df['Reading Score'] >= 70, :].groupby("School Name")
read_counts = school_read['Student ID'].count()
read_rates = (read_counts/students_by_school)*100
# Overall Passing Rates
overall_pass_rates = (math_rates + read_rates)/2
```

```python
School_Summary_df = pd.DataFrame({'School Type': school_type, 'Total Students': students_by_school, 'Total School Budget': school_budget_by_school, 
                               'Per Student Budget': student_budget_by_school, 'Average Math Score': average_math_score_by, 
                               'Average Reading Score': average_reading_score_by, '% Passing Math': math_rates, 
                               '% Passing Reading': read_rates, 'Overall Passing Rate': overall_pass_rates})
# Display the data frame
School_Summary_df
```

```python
top_schools_df = School_Summary_df.sort_values("Overall Passing Rate", ascending = False)
top_schools_df.head()
```

## Bottom Performing Schools (By Passing Rate)

* Sort and display the five worst-performing schools

```python
Bottom_schools_df = School_Summary_df.sort_values("Overall Passing Rate", ascending = True)
Bottom_schools_df.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Create the bins in which Data will be held
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]

# Slice the data and place it into bins
School_Summary_df["Spending (Per Student)"] = pd.cut(student_budget_by_school, spending_bins, labels = group_names)

## Creating a groups based off of the bins
spend_math_score = School_Summary_df.groupby(["Spending (Per Student)"]).mean()['Average Math Score']
spend_reading_score = School_Summary_df.groupby(["Spending (Per Student)"]).mean()['Average Reading Score']
spend_passing_math =  School_Summary_df.groupby(["Spending (Per Student)"]).mean()['% Passing Math']
spend_passing_reading =  School_Summary_df.groupby(["Spending (Per Student)"]).mean()['% Passing Reading']
overall_passing_rate =  (spend_passing_math + spend_passing_reading) / 2


# Establish Spending Dataframe
spend_df = pd.DataFrame({"Average Math Score":spend_math_score, "Average Reading Score":spend_reading_score,
                               "% Passing Math":spend_passing_math,"% Passing Reading":spend_passing_reading,
                                    "Overall Passing Rate":overall_passing_rate}) 
# Display data frame
spend_df
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
# Create the bins in which Data will be held
size_bins = [0, 1000, 2000, 5000]
size_names = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Categorize spending based on spending bins
School_Summary_df["School Size"] = pd.cut(students_by_school, size_bins, labels = size_names)

# Creating a groups based off of the bins
size_math_score = School_Summary_df.groupby(["School Size"]).mean()['Average Math Score']
size_reading_score = School_Summary_df.groupby(["School Size"]).mean()['Average Reading Score']
size_passing_math =  School_Summary_df.groupby(["School Size"]).mean()['% Passing Math']
size_passing_reading =  School_Summary_df.groupby(["School Size"]).mean()['% Passing Reading']
overall_passing_rate =  (size_math_score + size_reading_score) / 2


# Establish Spending Dataframe
size_df = pd.DataFrame({"Average Math Score":size_math_score, "Average Reading Score":size_reading_score,
                               "% Passing Math":size_passing_math,"% Passing Reading":size_passing_reading,
                                    "Overall Passing Rate":overall_passing_rate}) 
# Display data frame
size_df
```
